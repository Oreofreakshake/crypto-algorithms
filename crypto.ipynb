```python
import sympy as sp

def gcd_poly_over_field(poly1, poly2, field):
    F = sp.GF(field)

    # Convert polynomials to finite field representations
    poly1_F = sp.poly(poly1, domain=F)
    poly2_F = sp.poly(poly2, domain=F)

    # Compute the gcd
    gcd_poly = sp.gcd(poly1_F, poly2_F)

    # Return the simplified polynomial as a string or expression
    return gcd_poly.as_expr()

def main():
    x = sp.symbols('x')

    # first question
    poly_a1 = x**3 + 1
    poly_a2 = x**2 + x + 1
    gcd_a = gcd_poly_over_field(poly_a1, poly_a2, 2)
    print(f"Problem a: GCD(x^3 + 1, x^2 + x + 1) over GF(2) = {gcd_a}")

    # second question
    poly_b1 = x**3 + x + 1
    poly_b2 = x**2 + 1
    gcd_b = gcd_poly_over_field(poly_b1, poly_b2, 3)
    print(f"Problem b: GCD(x^3 + x + 1, x^2 + 1) over GF(3) = {gcd_b}")

    # third question
    poly_c1 = x**3 - 2*x + 1
    poly_c2 = x**2 - x - 2
    gcd_c = gcd_poly_over_field(poly_c1, poly_c2, 5)
    print(f"Problem c: GCD(x^3 - 2x + 1, x^2 - x - 2) over GF(5) = {gcd_c}")

    # fourth question
    poly_d1 = x**3 + 8*x**2 + 7*x + 8
    poly_d2 = 2*x**3 + 9*x**2 + 10*x + 1
    gcd_d = gcd_poly_over_field(poly_d1, poly_d2, 11)
    print(f"Problem d: GCD(x^3 + 8x^2 + 7x + 8, 2x^3 + 9x^2 + 10x + 1) over GF(11) = {gcd_d}")

if __name__ == "__main__":
    main()

```

Hill cypher algorithm **DECRYPTION**

```python
import numpy as np

def text_to_numbers(text):
    """Convert text to numbers (A=0, B=1, ..., Z=25)"""
    return [ord(c) - ord('A') for c in text.upper()]

def numbers_to_text(numbers):
    """Convert numbers back to text"""
    return ''.join(chr(n % 26 + ord('A')) for n in numbers)

def mod_inverse(a, m):
    """Find modular multiplicative inverse using extended Euclidean algorithm"""
    def extended_gcd(a, b):
        if a == 0:
            return b, 0, 1
        gcd, x1, y1 = extended_gcd(b % a, a)
        x = y1 - (b // a) * x1
        y = x1
        return gcd, x, y

    gcd, x, _ = extended_gcd(a % m, m)
    if gcd != 1:
        return None
    return x % m

def decrypt_message(ciphertext, key_matrix):
    print("\n=== DECRYPTION STEPS ===")

    numbers = text_to_numbers(ciphertext)
    print(f"\nCiphertext '{ciphertext}' as numbers:")
    print(numbers)

    triples = [numbers[i:i+3] for i in range(0, len(numbers), 3)]
    print("\nSplit into groups of three:")
    for i, triple in enumerate(triples, 1):
        while len(triple) < 3:  # Pad with zeros if needed
            triple.append(0)
        print(f"Group {i}: {triple}")

    K = np.array(key_matrix, dtype=int)
    print("\nKey Matrix K:")
    for row in K:
        print([f"{x:4}" for x in row])

    # Calculate determinant and its mod 26
    det = int(round(np.linalg.det(K)))
    det_mod26 = det % 26
    print(f"\nDeterminant of K: {det}")
    print(f"Determinant mod 26: {det_mod26}")

    # Find modular multiplicative inverse of determinant
    det_inv = mod_inverse(det_mod26, 26)
    if det_inv is None:
        raise ValueError("Matrix is not invertible modulo 26")
    print(f"Modular multiplicative inverse of determinant: {det_inv}")

    # Calculate adjugate matrix
    adj = np.zeros((3,3), dtype=int)
    for i in range(3):
        for j in range(3):
            minor = np.delete(np.delete(K, i, 0), j, 1)
            cofactor = int(round(np.linalg.det(minor))) * (-1)**(i+j)
            adj[j,i] = cofactor

    print("\nAdjugate Matrix (before mod 26):")
    print(adj)

    adj_mod26 = adj % 26
    print("\nAdjugate Matrix (after mod 26):")
    print(adj_mod26)

    # Calculate K⁻¹ = (adj_mod26 * det_inv) mod 26
    K_inv = (adj_mod26 * det_inv) % 26
    print("\nInverse Matrix K⁻¹ = (adj_mod26 * det_inv) mod 26:")
    print(K_inv)

    decrypted = []
    for triple in triples:
        print(f"\nProcessing triple: {triple}")
        result = np.dot(K_inv, triple) % 26
        print(f"After multiplication with K⁻¹: {result}")
        decrypted.extend(result)

    decrypted = decrypted[:len(ciphertext)]
    print("\nDecrypted numbers:", decrypted)

    plaintext = numbers_to_text(decrypted)
    return plaintext

K = [
    [2, 4, 12],
    [1, 2, 6],
    [3, 6, 18]
]


ciphertext = "KSCECG"

print("=== 3x3 HILL CIPHER DECRYPTION ===")
print(f"\nOriginal ciphertext: {ciphertext}")
try:
    plaintext = decrypt_message(ciphertext, K)
    print(f"\nDecrypted message: {plaintext}")
except ValueError as e:
    print(f"\nError: {e}")

def verify_key_matrix(K):
    """Verify if the key matrix is valid for Hill cipher"""
    K = np.array(K, dtype=int)
    det = int(round(np.linalg.det(K))) % 26
    if det == 0:
        return False, "Matrix is singular (determinant = 0 mod 26)"
    if mod_inverse(det, 26) is None:
        return False, "Determinant has no multiplicative inverse mod 26"
    return True, "Matrix is valid for Hill cipher"

valid, message = verify_key_matrix(K)
print(f"\nKey Matrix Verification: {message}")
```

```python
import numpy as np

def text_to_numbers(text):
    return [ord(c) - ord('A') for c in text]

def numbers_to_text(numbers):
    return ''.join(chr(n + ord('A')) for n in numbers)

def process_group(K, group):
    group = np.array(group)
    dot_product = np.dot(K, group)
    result = dot_product % 26
    return result.astype(int)

def encrypt_message(plaintext, key_matrix):
    print("\n=== ENCRYPTION STEPS ===")

    numbers = text_to_numbers(plaintext)
    print(f"\nPlaintext '{plaintext}' as numbers:")
    print(numbers)

    K = np.array(key_matrix)
    rows, cols = K.shape
    print(f"\nKey matrix dimensions: {rows}x{cols}")

    groups = []
    for i in range(0, len(numbers), cols):
        group = numbers[i:i+cols]
        if len(group) < cols:
            group.extend([0] * (cols - len(group)))  # Pad with zeros
        groups.append(group)
    print("\nSplit into groups:")
    for i, group in enumerate(groups, 1):
        letters = ''.join(chr(n + ord('A')) for n in group)
        print(f"Group {i}: {letters} -> {group}")

    numbers_flat = []
    for group in groups:
        numbers_flat.extend(group)

    print("\nKey Matrix K:")
    print(K)

    encrypted = []
    for i in range(0, len(numbers_flat), cols):
        group = numbers_flat[i:i+cols]
        letters = ''.join(chr(n + ord('A')) for n in group)
        print(f"\nProcessing group {i//cols + 1}:")
        print(f"Input letters: {letters}")
        print(f"Input numbers: {group}")

        result = process_group(K, group)
        result_letters = ''.join(chr(n + ord('A')) for n in result)
        print(f"Output numbers: {result}")
        print(f"Output letters: {result_letters}")

        remaining = min(rows, len(plaintext) - i)
        encrypted.extend(result[:remaining])

    print("\nAll encrypted numbers:", encrypted)

    ciphertext = numbers_to_text(encrypted[:len(plaintext)])
    return ciphertext

# Test cases
print("\nTest with 4x4 matrix:")
K_4x4 = [
    [3, 10, 20, 11],
    [5, 7, 14, 2],
    [9, 4, 8, 21],
    [11, 3, 17, 6]
]
plaintext = "ENCRYPTION"
print("=== MATRIX ENCRYPTION ===")
print(f"\nOriginal plaintext: {plaintext}")
ciphertext = encrypt_message(plaintext, K_4x4)
print(f"\nEncrypted message: {ciphertext}")

print("\nTest with 3x3 matrix:")
K_3x3 = [
    [2, 4, 12],
    [1, 2, 6],
    [3, 6, 18]
]
plaintext = "TESTIN"
print("=== MATRIX ENCRYPTION ===")
print(f"\nOriginal plaintext: {plaintext}")
ciphertext = encrypt_message(plaintext, K_3x3)
print(f"\nEncrypted message: {ciphertext}")

print("\nTest with 2x2 matrix:")
K_2x2 = [
    [7, 3],
    [2, 5]
]
plaintext = "meetmeattheusualplaceattenratherthaneightoclock"
print("=== MATRIX ENCRYPTION ===")
print(f"\nOriginal plaintext: {plaintext}")
ciphertext = encrypt_message(plaintext, K_2x2)
print(f"\nEncrypted message: {ciphertext}")
```

Determinant of mod 26

```python
def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = extended_gcd(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y

def mod_inverse(a, m):
    gcd, x, _ = extended_gcd(a, m)
    if gcd != 1:
        return None  # Modular inverse doesn't exist
    return x % m

def get_2x2_determinant(matrix):
    return (matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]) % 26

def get_minor_matrix(matrix, i, j):
    return [[matrix[row][col] for col in range(len(matrix)) if col != j]
            for row in range(len(matrix)) if row != i]

def determinant_mod26(matrix):
    size = len(matrix)

    if size == 2:
        return get_2x2_determinant(matrix)
    elif size == 3:
        det = 0
        # Use first row for cofactor expansion
        for j in range(3):
            minor = get_minor_matrix(matrix, 0, j)
            cofactor = (-1) ** j * matrix[0][j] * get_2x2_determinant(minor)
            det = (det + cofactor) % 26
        return det
    else:
        raise ValueError("Matrix must be 2x2 or 3x3")

def print_matrix(matrix):
    for row in matrix:
        print([f"{x:2d}" for x in row])

def test_determinant(matrix):
    size = len(matrix)
    print(f"\nCalculating determinant for {size}x{size} matrix:")
    print("\nInput Matrix:")
    print_matrix(matrix)

    if size == 2:
        a, b = matrix[0][0], matrix[0][1]
        c, d = matrix[1][0], matrix[1][1]
        print(f"\nStep-by-step calculation:")
        print(f"1. For 2x2 matrix, det = ad - bc")
        print(f"2. det = ({a} × {d}) - ({b} × {c})")
        print(f"3. det = {a*d} - {b*c}")
        print(f"4. det = {a*d - b*c}")
        det = determinant_mod26(matrix)
        print(f"5. det mod 26 = {det}")

    else:  # 3x3 matrix
        print("\nStep-by-step calculation using cofactor expansion along first row:")
        det = 0
        for j in range(3):
            minor = get_minor_matrix(matrix, 0, j)
            print(f"\nCofactor ({0+1},{j+1}):")
            print(f"Element: {matrix[0][j]}")
            print("Minor matrix:")
            print_matrix(minor)
            minor_det = get_2x2_determinant(minor)
            cofactor = (-1) ** j * matrix[0][j] * minor_det
            print(f"Minor determinant: {minor_det}")
            print(f"Cofactor calculation: ({-1 if j%2 else 1}) × {matrix[0][j]} × {minor_det} = {cofactor}")
            det = (det + cofactor) % 26
            print(f"Running sum (mod 26): {det}")

        print(f"\nFinal determinant (mod 26): {det}")

    # Check if determinant is valid
    print(f"\nIs determinant valid (has multiplicative inverse mod 26)? {mod_inverse(det, 26) is not None}")
    inv = mod_inverse(det, 26)
    if inv is not None:
        print(f"Multiplicative inverse of determinant mod 26: {inv}")

# Test cases
def run_tests():
    # 2x2 matrix test
    matrix_2x2 = [
        [23, 5],
        [13, 7]
    ]

    # 3x3 matrix test
    matrix_3x3 = [
        [21, 13, 25],
        [5, 7, 18],
        [3, 14, 12]
    ]

    # Test both matrices
    test_determinant(matrix_2x2)
    test_determinant(matrix_3x3)

if __name__ == "__main__":
    run_tests()
```

Inverse of mod 26

```python
def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = extended_gcd(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y

def mod_inverse(a, m):
    gcd, x, _ = extended_gcd(a, m)
    if gcd != 1:
        return None
    return x % m

def get_2x2_determinant(matrix):
    return (matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]) % 26

def get_minor_matrix(matrix, i, j):
    return [[matrix[row][col] for col in range(len(matrix)) if col != j]
            for row in range(len(matrix)) if row != i]

def get_3x3_determinant(matrix):
    det = 0
    for j in range(3):
        minor = get_minor_matrix(matrix, 0, j)
        cofactor = (-1) ** j * matrix[0][j] * get_2x2_determinant(minor)
        det = (det + cofactor) % 26
    return det

def get_cofactor_matrix(matrix):
    size = len(matrix)
    cofactor_matrix = [[0 for _ in range(size)] for _ in range(size)]

    for i in range(size):
        for j in range(size):
            minor = get_minor_matrix(matrix, i, j)
            if size == 2:
                cofactor_matrix[i][j] = ((-1) ** (i + j) * minor[0][0]) % 26
            else:  # 3x3
                cofactor_matrix[i][j] = ((-1) ** (i + j) * get_2x2_determinant(minor)) % 26

    return cofactor_matrix

def transpose_matrix(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix))]

def matrix_inverse_mod26(matrix):
    size = len(matrix)
    if size not in [2, 3]:
        raise ValueError("Matrix must be 2x2 or 3x3")

    # Calculate determinant
    det = get_2x2_determinant(matrix) if size == 2 else get_3x3_determinant(matrix)

    # Find modular multiplicative inverse of determinant
    det_inverse = mod_inverse(det, 26)
    if det_inverse is None:
        return None

    # Calculate cofactor matrix
    cofactor_matrix = get_cofactor_matrix(matrix)

    # Get adjugate matrix (transpose of cofactor matrix)
    adjugate = transpose_matrix(cofactor_matrix)

    # Multiply adjugate by determinant inverse modulo 26
    inverse = [[((adjugate[i][j] * det_inverse) % 26) for j in range(size)]
               for i in range(size)]

    return inverse

def print_matrix(matrix, label="Matrix"):
    print(f"\n{label}:")
    for row in matrix:
        print([f"{x:2d}" for x in row])

def test_matrix_inverse(matrix):
    size = len(matrix)
    print(f"\nCalculating inverse for {size}x{size} matrix:")
    print_matrix(matrix, "Input Matrix")

    det = get_2x2_determinant(matrix) if size == 2 else get_3x3_determinant(matrix)
    print(f"\nStep 1: Calculate determinant")
    print(f"Determinant mod 26 = {det}")

    det_inverse = mod_inverse(det, 26)
    print(f"\nStep 2: Calculate determinant inverse mod 26")
    if det_inverse is None:
        print("Matrix is not invertible (determinant has no inverse mod 26)")
        return
    print(f"Determinant inverse = {det_inverse}")

    cofactor_matrix = get_cofactor_matrix(matrix)
    print("\nStep 3: Calculate cofactor matrix")
    print_matrix(cofactor_matrix, "Cofactor Matrix")

    adjugate = transpose_matrix(cofactor_matrix)
    print("\nStep 4: Calculate adjugate (transpose of cofactor matrix)")
    print_matrix(adjugate, "Adjugate Matrix")

    inverse = matrix_inverse_mod26(matrix)
    print("\nStep 5: Calculate inverse (adjugate × determinant inverse mod 26)")
    print_matrix(inverse, "Inverse Matrix")

    print("\nVerification:")
    print("Multiply original matrix by inverse (should give identity matrix mod 26)")
    result = [[sum(matrix[i][k] * inverse[k][j] for k in range(size)) % 26
               for j in range(size)] for i in range(size)]
    print_matrix(result, "Result of multiplication")

def run_tests():
    matrix_2x2 = [
        [2, 3],
        [1, 22]
    ]

    matrix_3x3 = [
        [6, 24, 1],
        [13, 16, 10],
        [20, 17, 15]
    ]

    print("\n=== Testing 2x2 Matrix ===")
    test_matrix_inverse(matrix_2x2)

    print("\n=== Testing 3x3 Matrix ===")
    test_matrix_inverse(matrix_3x3)

if __name__ == "__main__":
    run_tests()
```

```python
def decrypt_playfair():
    original_ciphertext = "KXJEY UREBE ZWEHE WRYTU HEYFS KREHE GOYFI WTTTU OLKSY CAJPO BOTEI ZONTX BYBNT GONEY CUZWR GDSON SXBOU YWRHE BAAHY USEDQ"

    ciphertext = original_ciphertext.replace(" ", "")

    # Key: "ROYAL NEW ZEALAND NAVY"
    def create_matrix(key):

        key = key.replace(" ", "").upper()
        key = key.replace("J", "I")
        seen = set()
        key = "".join([c for c in key if not (c in seen or seen.add(c))])
        alphabet = "".join(c for c in "ABCDEFGHIKLMNOPQRSTUVWXYZ" if c not in key)

        matrix_chars = key + alphabet

        return [list(matrix_chars[i:i+5]) for i in range(0, 25, 5)]

    def print_matrix(matrix):
        print("\nPlayfair Matrix:")
        for row in matrix:
            print(" ".join(row))
        print()

    def find_position(matrix, char):
        char = 'I' if char == 'J' else char
        for i in range(5):
            for j in range(5):
                if matrix[i][j] == char:
                    return i, j
        return None

    def decrypt_pair(matrix, pair):
        row1, col1 = find_position(matrix, pair[0])
        row2, col2 = find_position(matrix, pair[1])

        if row1 == row2:
            return matrix[row1][(col1 - 1) % 5] + matrix[row2][(col2 - 1) % 5]
        elif col1 == col2:
            return matrix[(row1 - 1) % 5][col1] + matrix[(row2 - 1) % 5][col2]
        else:
            return matrix[row1][col2] + matrix[row2][col1]

    matrix = create_matrix("ROYAL NEW ZEALAND NAVY")
    print("Original ciphertext:")
    print(original_ciphertext)
    print_matrix(matrix)

    pairs = [ciphertext[i:i+2] for i in range(0, len(ciphertext), 2)]
    print("Processing digraphs:")

    decrypted_pairs = []
    for pair in pairs:
        if len(pair) == 2:
            decrypted = decrypt_pair(matrix, pair)
            decrypted_pairs.append(decrypted)
            print(f"{pair} → {decrypted}")

    plaintext = "".join(decrypted_pairs)

    formatted = " ".join([plaintext[i:i+5] for i in range(0, len(plaintext), 5)])

    print("\nFinal plaintext (in groups of 5):")
    print(formatted)

    return plaintext

decrypted_text = decrypt_playfair()
```

```python
def playfair_encrypt():
    plaintext = "I ONLY REGRET THAT I HAVE BUT ONE LIFE TO GIVE FOR MY COUNTRY"

    matrix = [
        ['J/K', 'C', 'D', 'E', 'F'],
        ['U', 'N', 'P', 'Q', 'S'],
        ['Z', 'V', 'W', 'X', 'Y'],
        ['R', 'A', 'L', 'G', 'O'],
        ['B', 'I', 'T', 'H', 'M']
    ]

    def print_matrix(matrix):
        print("\nPlayfair Matrix:")
        for row in matrix:
            print(" ".join(row))
        print()

    def find_position(matrix, char):
        """Find position of a character in the matrix."""
        for i in range(5):
            for j in range(5):
                if char in matrix[i][j]:  # Check if char is in cell (handles J/K case)
                    return i, j
        return None

    def prepare_text(text):
        """Prepare text for Playfair encryption."""
        text = text.upper().replace(" ", "")
        # Handle J and K (they share a cell)
        text = text.replace("J", "I/K").replace("K", "I/K")

        prepared = []
        i = 0
        while i < len(text):
            if i == len(text) - 1:
                prepared.append(text[i] + "X")
                break
            elif text[i] == text[i + 1]:
                prepared.append(text[i] + "X")
                i += 1
            else:
                prepared.append(text[i] + text[i + 1])
                i += 2

        return prepared

    def encrypt_pair(matrix, pair):
        """Encrypt a pair of characters using Playfair rules."""
        row1, col1 = find_position(matrix, pair[0])
        row2, col2 = find_position(matrix, pair[1])

        if row1 == row2:
            return matrix[row1][(col1 + 1) % 5] + matrix[row2][(col2 + 1) % 5]
        elif col1 == col2:
            return matrix[(row1 + 1) % 5][col1] + matrix[(row2 + 1) % 5][col2]
        else:
            return matrix[row1][col2] + matrix[row2][col1]

    print("Original plaintext:")
    print(plaintext)
    print_matrix(matrix)

    pairs = prepare_text(plaintext)
    print("Prepared text in digraphs:")
    print(" ".join(pairs))

    encrypted_pairs = []
    print("\nEncryption process:")
    for pair in pairs:
        encrypted = encrypt_pair(matrix, pair)
        encrypted_pairs.append(encrypted)
        print(f"{pair} → {encrypted}")

    ciphertext = "".join(encrypted_pairs)

    formatted = " ".join([ciphertext[i:i+5] for i in range(0, len(ciphertext), 5)])

    print("\nFinal ciphertext (in groups of 5):")
    print(formatted)

    return ciphertext

encrypted_text = playfair_encrypt()
```
